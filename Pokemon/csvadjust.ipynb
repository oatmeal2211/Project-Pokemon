```python
import re

data = """
 Level 	 Move 	 Type 	 Power 	 Accuracy 	 PP 
1	Gust	Normal	40	100%	35
5	Sand-Attack	Normal	—	100%	15
12	Quick Attack	Normal	40	100%	30
19	Whirlwind	Normal	—	85%	20
28	Wing Attack	Flying	35	100%	35
36	Agility	Psychic	—	—%	30
44	Mirror Move	Flying	—	—%	20
"""
id = 6
# 使用正则表达式提取Level,Move,Type
pattern = re.compile(r'(\d+)\s+([A-Za-z]+(?:\s+[A-Za-z]+)*)\s+([A-Za-z]+(?:\s+[A-Za-z]+)*)')
matches = pattern.findall(data)

# 打印结果
# for match in matches:
#     print(f"Level: {match[0]}, Move: {match[1]}")
# print(matches)



```

```python

for i, (level, move, type_) in enumerate(matches):
    df.loc[df['id'] == id, f'move{i+1}'] = move
    df.loc[df['id'] == id, f'type{i+1}'] = type_
    df.loc[df['id'] == id, f'need_level{i+1}'] = level

# 显示更新后的DataFrame
print(df)
```

```python
df.to_csv('Combine.csv', index=False)
```
